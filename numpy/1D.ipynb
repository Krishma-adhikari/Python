# Arithmetic Operations

```python
a = np.array([1,4,3])
b = np.array([2,4,3])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a % b)
print(a @ b)
```

# Statistical operations

```python
print(np.max(a))
print(np.min(a))
print(np.sum(a))
print(np.var(a))
print(np.mean(a))
print(np.median(a))
print(np.std(a))
```
